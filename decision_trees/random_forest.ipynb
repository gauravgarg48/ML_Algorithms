## Random Forests

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load Data
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_train.head(n=10)
```

```python
# Clean Train and test data

inp_cols = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']
out_col = ['survived']

features_to_drop = ['name', 'ticket', 'cabin', 'embarked', 'boat', 'body', 'home.dest']
df_train = df_train.drop(columns=features_to_drop)
df_test = df_test.drop(columns=features_to_drop)

ua = df_train['age'].mean()
uf = df_train['fare'].mean()

values = {'age': ua, 'fare': uf}
df_train.fillna(values, inplace=True)
df_test.fillna(values, inplace=True)

le = LabelEncoder()
df_train['sex'] = le.fit_transform(df_train['sex'])
df_test['sex'] = le.transform(df_test['sex'])

df_train.info()
```

```python
X = np.array(df_train[inp_cols])
y = np.array(df_train[out_col]).reshape((-1,))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Train decision tree

dt_clf = DecisionTreeClassifier(criterion='entropy')
dt_clf.fit(X_train, y_train)

# Training accuracy
print(dt_clf.score(X_train, y_train))

# Testing accuracy
print(dt_clf.score(X_test, y_test))
```

## Train Random forest and check if it overfits

```python
rf_clf = RandomForestClassifier(n_estimators=15, criterion='entropy', max_depth=5)
rf_clf.fit(X_train, y_train)

print(rf_clf.score(X_train, y_train))
print(rf_clf.score(X_test, y_test))
```

```python
# Tune Hyperparameter- n_estimators

acc_list = []
for i in range(1,50):
    acc = cross_val_score(RandomForestClassifier(n_estimators=i,max_depth=5),
                          X_train, y_train, cv=5).mean()
    acc_list.append(acc)

plt.plot(acc_list)
plt.grid('on')
plt.show()
```

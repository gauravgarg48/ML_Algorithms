## K-means vs DBSCAN

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles, make_moons
from sklearn.cluster import KMeans, DBSCAN
```

```python
# Plot data

def plotData(X, labels, centers=None):
    
    plt.scatter(X[:, 0], X[:, 1], c=labels)
    
    if centers is not None:
        plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='*')
    
    plt.show()
```

```python
# Load moons dataset

X, y = make_moons(n_samples=300, noise=0.1)
print(X.shape, y.shape)

# Visualize dataset

plotData(X, y)
```

```python
km = KMeans(n_clusters=2)
km.fit(X)
```

```python
plotData(X, km.labels_, km.cluster_centers_)
```

# DBSCAN
### Density Based Spatial Clustering of Applications with Noise

```python
dbs = DBSCAN(eps = 0.18, min_samples = 5)
y_pred = dbs.fit_predict(X)

plotData(X, y_pred)
```

```python
# Load circles dataset

X, y = make_circles(n_samples=400, noise=0.02)
print(X.shape, y.shape)

plotData(X, y)
```

```python
dbs = DBSCAN(eps = 0.1, min_samples = 4)
y_pred = dbs.fit_predict(X)

plotData(X, y_pred)
```

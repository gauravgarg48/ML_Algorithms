## Implementation of Gradient Descent (in general)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(100)
Y = (X-50)**2

print(X)
print(Y)
```

```python
# Visualization

plt.style.use("seaborn")
plt.plot(X,Y)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

```python
x = 0
lr = 0.1
error = []
plt.plot(X, Y)

# 50 steps in the downhill direction
for i in range(50):
    grad = 2*(x-50)
    x = x - lr*grad
    y = (x-50)**2
    error.append(y)
    plt.scatter(x, y)
```

```python
# Plot the error

plt.plot(error)
plt.show()
```
